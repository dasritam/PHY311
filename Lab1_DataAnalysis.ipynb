```python
import numpy as np
import imageio
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
N = 8
def generate(N, d):
    pic = imageio.imread('images/image'+str(N)+'.jpg')
    plt.figure(figsize = (5,5))
    plt.imshow(pic)
    plt.savefig('results/z'+str(N)+'_original.jpg', dpi=300)
    plt.show()
    p = pic[:d,:d,:]
    im = p[:,:,0]+p[:,:,1]+p[:,:,2]
    plt.figure(figsize = (10,10))
    plt.imshow(im, cmap='viridis')
    plt.savefig('results/z'+str(N)+'_scary.jpg', dpi=300)
    plt.show()

    x = np.linspace(0, 255, d)
    x, y = np.meshgrid(x, x)
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, im, cmap="magma")
    plt.savefig('results/z'+str(N)+'_surface.jpg', dpi=300)
    plt.show()
    
n = [3,5,6,8]
D = [560,1250,1200,700]

for i in range(4):
    generate(n[i],D[i])
```
